```python
dire='time_rest_0.06'
```

```python
ligo_overtone=np.loadtxt('../time_rest/epsilong_overtone_join')
ligo_time_overtone=np.loadtxt('../time_rest/tinit_overtone_join')

inj_overtone=np.loadtxt(dire+'/epsilong_overtone_join')
inj_time_overtone=np.loadtxt(dire+'/tinit_overtone_join')

ligo=np.loadtxt('../time_rest/epsilong_join')
ligo_time=np.loadtxt('../time_rest/tinit_join')

inj=np.loadtxt(dire+'/epsilong_join')
inj_time=np.loadtxt(dire+'/tinit_join')
```

```python
t_unit=68.5*2950./2/299792458
mean_mass=(34+140)/2
mean_spin=(0.1+0.95)/2
mean_epsilong=np.sqrt(((mean_mass-68.5)/68.5)**2+(mean_spin-0.692085186818)**2)
```

```python
fig = pl.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()

ax1.plot(inj_time,inj,'o',ms=2,color='r')
ax1.plot(inj_time,inj,'--',lw=0.2,color='r')

ax1.plot(ligo_time,ligo,'o',ms=2,color='b')
ax1.plot(ligo_time,ligo,'--',lw=0.2,color='b')

ax2.plot(ligo_time/t_unit*1e-3,ligo,'o',ms=2,color='b')
# ax2.set_xlim(-5,30)
# ax1.set_xlim(-5*t_unit*1e3,30*t_unit*1e3)
ax1.set_xlabel(r'$\Delta t\, ({\rm ms})$')
ax2.set_xlabel(r'$\Delta t\, (M_f^{\rm IMR})$',labelpad=10)
ax1.set_ylabel(r'$\epsilon$')

ax1.axhline(y=mean_epsilong,ls='--',color='k')
# pl.axhline(y=0.69,lw=1)
# pl.savefig('GW150914_likelihood_time.pdf',bbox_inches='tight')
```

```python
fig = pl.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()

ax1.plot(inj_time_overtone,inj_overtone,'o',ms=2,color='r')
ax1.plot(inj_time_overtone,inj_overtone,'--',lw=0.2,color='r')

ax1.plot(ligo_time_overtone,ligo_overtone,'o',ms=2,color='b')
ax1.plot(ligo_time_overtone,ligo_overtone,'--',lw=0.2,color='b')

ax2.plot(ligo_time_overtone/t_unit*1e-3,ligo_overtone,'o',ms=2,color='b')
# ax2.set_xlim(-5,30)
# ax1.set_xlim(-5*t_unit*1e3,30*t_unit*1e3)
ax1.set_xlabel(r'$\Delta t\, ({\rm ms})$')
ax2.set_xlabel(r'$\Delta t\, (M_f^{\rm IMR})$',labelpad=10)
ax1.set_ylabel(r'$\epsilon$')

ax1.axhline(y=mean_epsilong,ls='--',color='k')
# pl.axhline(y=0.69,lw=1)
# pl.savefig('GW150914_likelihood_time.pdf',bbox_inches='tight')
```

```python
def joinf(name):
    if name==1:
        suffix='_overtone'
        prefix=''
    elif name==2:
        suffix=''
        prefix=''
    elif name==3:
        suffix=''
        prefix='no_filter_'
        
    mtrue=68.5
    mass_max_overtone1=np.loadtxt('noise_time_rest/'+prefix+'mass1'+suffix)
    spin_max_overtone1=np.loadtxt('noise_time_rest/'+prefix+'spin1'+suffix)
    tinit_overtone1=np.loadtxt('noise_time_rest/'+prefix+'tinit1'+suffix)
    epsilong_overtone1=np.sqrt(((mass_max_overtone1-mtrue)/mtrue)**2+(spin_max_overtone1-0.69)**2)
    
    mass_max_overtone2=np.loadtxt('noise_time_rest/'+prefix+'mass2'+suffix)
    spin_max_overtone2=np.loadtxt('noise_time_rest/'+prefix+'spin2'+suffix)
    tinit_overtone2=np.loadtxt('noise_time_rest/'+prefix+'tinit2'+suffix)
    epsilong_overtone2=np.sqrt(((mass_max_overtone2-mtrue)/mtrue)**2+(spin_max_overtone2-0.69)**2)
    
    mass_max_overtone3=np.loadtxt('noise_time_rest/'+prefix+'mass3'+suffix)
    spin_max_overtone3=np.loadtxt('noise_time_rest/'+prefix+'spin3'+suffix)
    tinit_overtone3=np.loadtxt('noise_time_rest/'+prefix+'tinit3'+suffix)
    epsilong_overtone3=np.sqrt(((mass_max_overtone3-mtrue)/mtrue)**2+(spin_max_overtone3-0.69)**2)
    
    mass_max_overtone4=np.loadtxt('noise_time_rest/'+prefix+'mass4'+suffix)
    spin_max_overtone4=np.loadtxt('noise_time_rest/'+prefix+'spin4'+suffix)
    tinit_overtone4=np.loadtxt('noise_time_rest/'+prefix+'tinit4'+suffix)
    epsilong_overtone4=np.sqrt(((mass_max_overtone4-mtrue)/mtrue)**2+(spin_max_overtone4-0.69)**2)
    
    mass_max_overtone5=np.loadtxt('noise_time_rest/'+prefix+'mass5'+suffix)
    spin_max_overtone5=np.loadtxt('noise_time_rest/'+prefix+'spin5'+suffix)
    tinit_overtone5=np.loadtxt('noise_time_rest/'+prefix+'tinit5'+suffix)
    epsilong_overtone5=np.sqrt(((mass_max_overtone5-mtrue)/mtrue)**2+(spin_max_overtone5-0.69)**2)
    
    mass_max_overtone6=np.loadtxt('noise_time_rest/'+prefix+'mass6'+suffix)
    spin_max_overtone6=np.loadtxt('noise_time_rest/'+prefix+'spin6'+suffix)
    tinit_overtone6=np.loadtxt('noise_time_rest/'+prefix+'tinit6'+suffix)
    epsilong_overtone6=np.sqrt(((mass_max_overtone6-mtrue)/mtrue)**2+(spin_max_overtone6-0.69)**2)
    
    epsilong_overtone_join=np.concatenate((epsilong_overtone1,epsilong_overtone2,epsilong_overtone3,\
                                           epsilong_overtone4,epsilong_overtone5,epsilong_overtone6))
    tinit_overtone_join=np.concatenate((tinit_overtone1,tinit_overtone2,tinit_overtone3,\
                                       tinit_overtone4,tinit_overtone5,tinit_overtone6))
    mass_overtone_join=np.concatenate((mass_max_overtone1,mass_max_overtone2,mass_max_overtone3,\
                                      mass_max_overtone4,mass_max_overtone5,mass_max_overtone6))

    return epsilong_overtone_join,tinit_overtone_join,mass_overtone_join
```
